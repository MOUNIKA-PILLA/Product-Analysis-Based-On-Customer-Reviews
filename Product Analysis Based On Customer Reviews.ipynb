```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
polarity_scores = []
for i in range(0,len(detail_review)):
    polarity_scores.append(sia.polarity_scores(detail_review[i]))
polarity_scores
```

```python
scores = []
for i in range(0,len(polarity_scores)):
    val = polarity_scores[i]['compound']
    if(val<0):
        scores.append('Negative')
    elif(val>0.5):
        scores.append('Positive')
    else:
        scores.append('Neutral')
scores
```

```python
df['Analysis'] = scores
df
```
